### Regression plot

Plot data and a linear regression model fit

Finds the best linear regression line which helps to predict the amount of change in Y on one unit change in X

```python
tips = sns.load_dataset('tips')
tips.head()
```

Create a basic reg plot

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

Change the styling of reg plot, only line,only scatter,show/hide the ci,change the ci value,change n_boot

```python
sns.regplot(data=tips,x='total_bill',y='tip',
            color='red',
            marker='o')
```

```python
sns.regplot(data=tips,x='total_bill',y='tip',
            line_kws=dict(color='red',linestyle="--"),
            scatter_kws=dict(s=100,color='green',alpha=0.5))
#shaded area is ci which is 95% by default
```

```python
sns.regplot(data=tips,x='total_bill',y='tip',
            line_kws=dict(color='red',linestyle="--"),
            scatter_kws=dict(s=100,color='green',alpha=0.5),
            n_boot=500)
```

Statistical models in seaborn

```python
sns.regplot(data=tips,x='total_bill',y='tip',order=4)
#order is for polynomial
```

### Heat Map

for this,dataframe should be in matrix form

Pivot or Correlation can be used for this

```python
mart = pd.read_csv(r"/content/mart.csv")
mart.columns = mart.columns.str.lower()
mart.head(10)
```

```python
martPiv = mart.pivot_table(index='outlet_year',
                           columns='outlet_size',
                           values='sales')
martPiv
```

```python
sns.set_style('white')
plt.figure(figsize=(5,5))
sns.heatmap(martPiv,
            #square=True
            annot=True,
            fmt='.0f',
            annot_kws=dict(size=15,weight='bold'),
            linewidths=0.5,
            #linecolor='black',
            cmap='OrRd_r'
            )
#fmt is decimal values which we want to see
#annot populates the no
#annot_kws for no appearance
#cmap is for palette
```

Using correlation

```python
mart.corr()
```

```python
sns.heatmap(mart.corr(),
            vmin=-1,
            vmax=1,
            center=0)
```

```python
sns.heatmap(mart.corr(),
            vmin=-1,
            vmax=1,
            center=0,
            cmap='OrRd_r',
            annot=True,
            fmt='.1f',
            annot_kws=dict(size=15,weight='bold'),
            linecolor='black',
            linewidths=.5)
```

### Cluster Map

Plot a matrix dataset as a hierarchically-clustered heatmap

```python
sns.clustermap(martPiv,
              #  col_cluster=False,
               row_cluster=False,
               annot=True,
               fmt='.0f',
               z_score=1,
               standard_scale=0
               )
#there is no sequqnce rather cluster of similar data
```

```python
obj = sns.FacetGrid(mart,col='outlet_size')
```

```python
obj = sns.FacetGrid(mart,col='outlet_size')
obj.map_dataframe(sns.histplot,x='sales')
```

```python
obj = sns.FacetGrid(mart,col='outlet_size')
obj.map_dataframe(sns.kdeplot,x='sales')
```

```python
obj = sns.FacetGrid(mart,col='outlet_size')
obj.map_dataframe(sns.lineplot,x='outlet_year',y='sales')
```

```python
obj = sns.FacetGrid(mart,col='outlet_size')
obj.map_dataframe(sns.scatterplot,x='outlet_year',
                  y='sales',
                  hue='outlet_location_type',
                  marker='+',
                  alpha=0.5,
                  color='red',
                  s=100)
```

```python
obj = sns.FacetGrid(mart,col='outlet_size',row='outlet_location_type')
```

```python
obj = sns.FacetGrid(mart,col='outlet_size',row='outlet_location_type')
obj.map_dataframe(sns.histplot,x='sales')
obj.set_titles(col_template='{col_name} size',row_template='{row_name} Type')
```
