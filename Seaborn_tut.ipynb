### CountPlot

used for frequency of the categories

```python
sns.countplot(x='Gender',data=mart)
```

Create a basic Countplot to get number of transaction for each of each of product line

```python
sns.countplot(x='Product line',data=mart)
```

Change the figure size to clearly see the labels

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',data=mart)
```

Make it horizontal bar plot

```python
sns.countplot(y='Product line',data=mart)
```

Add hue to get the count on two categories i.e. Product line and Gender

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',hue="Gender",data=mart)
```

Use different color palette

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',hue="Gender",palette="Dark2",data=mart)
```

Change Style using facecolor,linewidth and edge color

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',data=mart,facecolor=(0,0,0,0),
              linewidth=5,edgecolor=sns.color_palette('dark',3))
```

Change default aggregation method using Estimation Parameter

```python
#by default it is mean
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,palette='twilight',
            estimator=sum)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,palette='twilight',
            estimator=np.median)
```

### Boxplot

This provides a summary of supplied data,which inlcudes the information like:

Mean

Median

Minimum

Maximum

1st Quartile

3rd Quartile

Outliers


Create a basic Box Plot on one numeric variable

```python
sns.boxplot(x='Total',data=mart)
#left most line is min
#left border of box is 1st quartile
#middle line is median
#mean is dot(hidden by default)
#right border of box is 3rd quartile
#right most line is max
#points after max or before min are outliers
```

```python
#change width
sns.boxplot(x='Total',data=mart,width=0.5)
```

```python
#make plot vertical
sns.boxplot(y='Total',data=mart)
```

Create a basic box plot on one numeric variable by a categorical variable

```python
sns.boxplot(x='Payment',y='Total',data=mart)
```

Create a basic Box plot on one numeric variable by twi categorical variable using hue attribute

```python
sns.boxplot(x='Payment',y='Total',hue='Gender',data=mart)
```

Add mean marker in the box plot using Showmeans

```python
sns.boxplot(x='Payment',y='Total',data=mart,showmeans=True)
```

Change its style using meanprops

```python
sns.boxplot(x='Payment',y='Total',data=mart,showmeans=True,meanprops={"marker":"o",
                                                                     "markerfacecolor":"white",
                                                                      "markersize":"5",
                                                                      "markeredgecolor":"black" 
                                                                      })
```

Make horizontal box plot

```python
sns.boxplot(y='Payment',x='Total',data=mart,showmeans=True)
```

Change Palette,Line Width etc

```python
sns.boxplot(x='Payment',y='Total',data=mart,showmeans=True,
            palette='CMRmap',
            linewidth=8
            )
```

Create box plot for Each of the Numeric Variable in dataframe

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=mart)
```

### Violinplot

Kernel is pdf function joined with its mirror img about x axis and then overall graph rotated to be symmetrical around y axis

The boxplot is inserted in between of kernel

The max leaf width of kernel is kde

Create a basic Violin Plot

```python
sns.violinplot(y="Total",data=mart)
```

Create a ViolinPlot on 2 categorical and one numeric variable and use split

```python
sns.violinplot(x='Payment',y='Total',data=mart) #1 categorical
```

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender") #2 categorical
```

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True)
```

Change the box in violinplot to horizontal lines

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True,inner='quartile')
```

Draw line for each observations in a ViolinPlot

```python
#the black lines show each single observations
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True,inner='stick')
```

Change the amount of smoothing of plot using bw attribute

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True,inner='quartile',bw=1.2)
```

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True,inner='quartile',bw=.2)
```

To cutout the extreme values

```python
sns.violinplot(x='Payment',y='Total',data=mart,hue="Gender",split=True,inner='quartile',bw=1.2,cut=0)
```

### Strip Plot

It helps in showing distinguished plot of repeated points

A strip plot is a graphical data analysis technique for summarzing a univariate dataset

A strip plot can be drawn on its own,but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution

Create a basic Strip Plot

```python
sns.stripplot(data=mart,y='Total')
```

Expand markers in Stripplot using jitter

```python
sns.stripplot(data=mart,y='Total',jitter=0.2)
```

Draw line around the points using Linewidth

```python
sns.stripplot(data=mart,y='Total',linewidth=0.5)
```

Include one categorical feature

```python
sns.stripplot(data=mart,y='Total',x='Payment')
```

```python
sns.stripplot(data=mart,y='Total',x='Payment',jitter=0.2,linewidth=0.2)
```

Including third categorical value with hue

```python
sns.stripplot(data=mart,y='Total',x='Payment',hue='Gender'
,jitter=0.2,linewidth=0.5)
```

Separate each level of hue using Dodge

```python
sns.stripplot(data=mart,y='Total',x='Payment',hue='Gender'
,jitter=0.2,linewidth=0.5,dodge=True)
```

Draw the strips on top of Violin Plot

```python
sns.stripplot(data=mart,y='Total',x='Payment')
sns.violinplot(data=mart,y='Total',x='Payment',color='gray')
```

Draw the strips on top of Box Plot

```python
sns.stripplot(data=mart,y='Total',x='Payment',color="black")
sns.boxplot(data=mart,y='Total',x='Payment',color='gray')
```
